```python
import numpy as np
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

display_labels = [0, 1, 2, 3]
final_cm = np.round((confusion_matrix(df['ans']-1, df['a']-1)+\
confusion_matrix(df['ans']-1, df['b']-1)+\
confusion_matrix(df['ans']-1, df['c']-1)+\
confusion_matrix(df['ans']-1, df['d']-1)+\
confusion_matrix(df['ans']-1, df['e']-1)+\
confusion_matrix(df['ans']-1, df['f']-1)+\
confusion_matrix(df['ans']-1, df['g']-1))/7, 2)


cm_display = ConfusionMatrixDisplay(confusion_matrix = final_cm, display_labels = display_labels)
cm_display.plot()
plt.title('Confusion Matrix')
plt.show()
```
